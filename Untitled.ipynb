```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.spatial import cKDTree

def generate_quadratic_bezier_curve(width, height, num=200):
    """
    幅 width, 高さ height の平面において
      P0 : 左下  (0, 0)
      P1 : 中央付近（ランダム、x ∈ [width/6, width*5/6], y ∈ [height/6, height*5/6]）
      P2 : 右上  (width, height)
    の3点から二次ベジェ曲線（スパイン）上の点列を返す。
    
    戻り値:
      curve : 二次ベジェ曲線上の点の numpy 配列 (num x 2)
      control_points : (P0, P1, P2) のタプル
      t_vals : 0～1 に対応するパラメータの配列（num,）
    """
    # 制御点の設定
    P0 = np.array([0, 0])
    P1 = np.array([
        random.uniform(width/6, width*5/6),
        random.uniform(height/6, height*5/6)
    ])
    P2 = np.array([width, height])
    
    t_vals = np.linspace(0, 1, num)
    # 二次ベジェ曲線： B(t) = (1-t)^2 P0 + 2(1-t)t P1 + t^2 P2
    curve = ((1-t_vals)**2)[:, np.newaxis] * P0 + 2 * ((1-t_vals)*t_vals)[:, np.newaxis] * P1 + (t_vals**2)[:, np.newaxis] * P2
    return curve, (P0, P1, P2), t_vals

def interpolate_color(c0, c1, t):
    """
    c0, c1 : それぞれRGBの3要素（0～1の範囲）
    t : 0～1 のスカラーまたは配列
    線形補間したRGB色を返す（スカラーもしくは配列の場合に対応）
    """
    c0 = np.array(c0)
    c1 = np.array(c1)
    return (1-t) * c0 + t * c1

def mosaic_random_rotate(img, tile_size=50, angle = 90):
    """
    画像を tile_size ピクセル四方のタイルに分割し、0,90,180,270 度のいずれかをランダムで回転して再配置する。
    端が割り切れない部分は四隅を切り落とす。
    
    Parameters
    ----------
    img : np.ndarray
        入力画像 (BGR 形式または RGB 形式でも可)。
    tile_size : int
        タイルの大きさ（画素）。デフォルトは 50。

    Returns
    -------
    np.ndarray
        タイルをランダム回転で再配置した画像。
        (img と同じチャンネル順になる。)
    """

    # 画像の高さ(H), 幅(W), チャンネル数(C)を取得
    h, w, c = img.shape

    # 端が余らないように切り落とす
    new_h = (h // tile_size) * tile_size
    new_w = (w // tile_size) * tile_size
    img_cropped = img[:new_h, :new_w]

    # 出力画像用に同じサイズの配列を用意
    mosaic_img = np.zeros_like(img_cropped)

    # 回転候補角度 (度数)
    angles = [0, 90, 180, 270][2:3]

    # タイル分割 & ランダム回転
    for y in range(0, new_h, tile_size):
        for x in range(0, new_w, tile_size):
            # タイルの切り出し
            tile = img_cropped[y:y+tile_size, x:x+tile_size]

            # angle に応じた回転
            # np.rot90: k=1 で 90度(反時計回り)回転、k=2 で 180度、k=3 で 270度
            if angle == 0:
                rotated_tile = tile
            elif angle == 90:
                rotated_tile = np.rot90(tile, k=1)
            elif angle == 180:
                rotated_tile = np.rot90(tile, k=2)
            elif angle == 270:
                rotated_tile = np.rot90(tile, k=3)

            # 回転後のタイルを mosaic_img に配置
            mosaic_img[y:y+tile_size, x:x+tile_size] = rotated_tile

    return mosaic_img

def main():
    # parameter
    width = 1600
    height = 900
    color_start = [1, 1, 0.7]
    color_end   = [1, 0.7, 1] 
    tile_size = 200

    # 曲線およびその制御点とパラメータtを取得
    num_curve_points = 200
    curve, control_points, t_vals = generate_quadratic_bezier_curve(width, height, num=num_curve_points)

    # 曲線上の各サンプルに対し、パラメータ t で色を補間した色を求める
    curve_colors = np.array([interpolate_color(color_start, color_end, t) for t in t_vals])  # shape: (num_curve_points, 3)

    # ここから、全ピクセルに対して「最近傍の曲線上の点」を求め、その色を埋める
    tree = cKDTree(curve)
    xs = np.arange(width)
    ys = np.arange(height)
    xv, yv = np.meshgrid(xs, ys)  # xv, yv shape: (height, width)
    pixel_coords = np.column_stack((xv.ravel(), yv.ravel())) # 各ピクセルの座標 (x,y) を (height*width, 2) に reshape
    _, idx = tree.query(pixel_coords)  # idx: (num_pixels,) 各ピクセルに対する曲線サンプルのインデックス
    pixel_colors = curve_colors[idx] # 各ピクセルの色は、曲線上の最近傍点の色とする
    image = pixel_colors.reshape((height, width, 3))# 画素画像へ変換

    # 回転処理
    image = mosaic_random_rotate(image, tile_size=tile_size) 
    
    # 結果を描画
    plt.figure(figsize=(8, 4.5))
    plt.imshow(image, origin='lower')   # origin='lower' により左下が (0,0) となる
    plt.axis('equal')
    plt.axis('off')
    plt.show()

if __name__ == '__main__':
    main()
```
